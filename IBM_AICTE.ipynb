```python
df=mdata
```

```python
df.info()
```

```python
#the country has the only column which has object datatype we have to convert it into numeric part

from sklearn.preprocessing import LabelEncoder
l=LabelEncoder()
for i in df.columns:
  if df[i].dtype == 'object':
    df[i]=l.fit_transform(df[i])
```

```python
x=df.drop('Mental-fitness',axis=1)
y=df['Mental-fitness']
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest= train_test_split(x,y,test_size=.20,random_state=2)
```

```python
#let us check the shape of train and test data
print("xtrain:",xtrain.shape)
print("xtest:",xtest.shape)
print("ytrain:",ytrain.shape)
print("ytest:",ytest.shape)

```

```python
from sklearn.metrics.pairwise import linear_kernel
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error,r2_score
lr=LinearRegression()
lr.fit(xtrain,ytrain)

ytrain_pred=lr.predict(xtrain)
mse= mean_squared_error(ytrain,ytrain_pred)

rootmse=(np.sqrt(mean_squared_error(ytrain,ytrain_pred)))

r2=r2_score(ytrain,ytrain_pred)


print("the result for training model is given below")
print("______________________________________________")
print("MSE is:", format(mse))
print("RMSE is:" , format(rootmse))
print("R2 is :", format(r2))
print("______________________________________________")
```

```python
from sklearn.ensemble import RandomForestRegressor
rfr=RandomForestRegressor()

rfr.fit(xtrain,ytrain)

ytrain_pred1=rfr.predict(xtrain)
mse=mean_squared_error(ytrain,ytrain_pred1)
rootmse1=(np.sqrt(mean_squared_error(ytrain,ytrain_pred1)))
r2=r2_score(ytrain,ytrain_pred)


print("The result for the Random Forest Regressor is given below")
print("__________________________________________________________")
print("MSE is:", format(mse))
print("RMSE is:" , format(rootmse1))
print("R2 is :", format(r2))
print("______________________________________________")
```

```python
ytest_pred=lr.predict(xtest)
mse= mean_squared_error(ytest,ytest_pred)

rootmse=(np.sqrt(mean_squared_error(ytest,ytest_pred)))

r2=r2_score(ytest,ytest_pred)
print("the result for training model is given below")
print("______________________________________________")
print("MSE is:", format(mse))
print("RMSE is:" , format(rootmse))
print("R2 is :", format(r2))
print("______________________________________________")
print("********************************************************************************************************************************************************************************")


ytest_pred1=rfr.predict(xtest)
mse=mean_squared_error(ytest,ytest_pred1)
rootmse1=(np.sqrt(mean_squared_error(ytest,ytest_pred1)))
r2=r2_score(ytest,ytest_pred)


print("The result for the Random Forest Regressor is given below")
print("__________________________________________________________")
print("MSE is:", format(mse))
print("RMSE is:" , format(rootmse1))
print("R2 is :", format(r2))
print("______________________________________________")
```
